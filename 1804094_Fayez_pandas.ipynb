```python
#Data frame like a query
a=df[df.temperature>=32]                                               #a query to print all the rows where temp>=32
print(a,"\n")
print(df[df.temperature==df.temperature.max()],"\n")                   # Query
b=df['day'][df.temperature==df['temperature'].max()]                   #a query to print all the rows where temp is max and column is day 
print(b,"\n")
```

```python
#Data frame index operation
print(df,"\n")
print(df.index,"\n")                              #current index
print(df.set_index('day'),"\n")                   #set day as an index. It doesn't modify the original frame
print(df,"\n")                                    #the original frame printed though i operate new index.
df.set_index('day',inplace=True)                  #set day as an index. It (inplace=true) modify the original frame
print(df,"\n")                                    #printed the modified frame
print(df.loc['1/3/2017'],"\n")                    #retrieve a specific row with the label '1/3/2017'.
df.reset_index(inplace=True)                      #back to the original index
print(df,'\n')                                    #printed the original frame
df.set_index('event',inplace=True)                #set event as an index. It (inplace=true) modify the original frame
print(df,"\n")                                    #printed the modified frame
print(df.loc['Rain'],"\n")                        #retrieve a specific row with the label 'Rain'.
df.reset_index(inplace=True)                      #back to the original index
```

#Handling Missing Data - fillna, interpolate, dropna

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/1804094_Fayez/Pandas/weather_data (1).csv")
df
```

```python
type(df.day[0])
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/1804094_Fayez/Pandas/weather_data (1).csv",parse_dates=['day'])
df
#parse_dates is an arguments. it change also data types.
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)  #inplace true change the original dataframe 
df
```

```python
new_df=df.fillna(0)                              #original frame will not change. Here all NaN value replace with 0.
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': 'not event'
})                                               #replace NaN with specific value for specific column
new_df
```

```python
new_df=df.fillna(method='ffill')  #ffill=forward fill. It replace NaN with previous value of entire frame
new_df
```

```python
new_df=df.fillna(method='bfill')  #ffill=backward fill. It replace NaN with next value of entire frame
new_df
```

```python
new_df=df.fillna(method='bfill',axis="columns")  #ffill=backward fill. It replace NaN with next value according to column of entire frame.By default it works according to rows 
new_df
```

```python
new_df=df.fillna(method='ffill',axis='columns')  #ffill=forward fill. It replace NaN with previous value according to column of entire frame
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1)  #ffill=forward fill. It replace NaN with previous value 1 time  of entire frame
new_df
```

```python
new_df=df.interpolate()                      #It replace NaN with a value between previous and next value of entire frame. By default it's linear.
new_df
#check interpolate documentation for more method
```

```python
new_df=df.interpolate(method='time') #date 4 and 5 are nearest. So the replacing value of NaN is nearest of date 4 and 5
new_df
```

```python
new_df=df.dropna()                            #Drop the row which has atleast one NaN value
new_df
```

```python
new_df=df.dropna(how='all')                                 #drop the rows which has all NaN value
new_df
```

```python
new_df=df.dropna(thresh=1)                                          #Keep the rows which has atleast 1 NaN value
new_df
```

```python
new_df=df.dropna(thresh=2)                            #Keep the rows which has atleast 2 NaN value. as date is index so it's not counted
new_df
```

```python
dt=pd.date_range('01-01-2017','01-11-2017')             #create the index of date 1 to 11. dt=date time
idx=pd.DatetimeIndex(dt) 
df=df.reindex(idx)
df
#we can fill the NaN value with fillna method.
```

#Pandas Merge Tutorial

```python
df1 = pd.DataFrame({
    "city": ["Dhaka","Chittagong","Khulna"],
    "temperature": [37,45,30],
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["Dhaka","Chittagong","Khulna"],
    "humidity": [80, 60, 78]
})
df2
```

```python
df3 = pd.merge(df1,df2, on ='city') #the column city which i want to merge 
df3
```

```python
#Type Of DataBase Joins
df1 = pd.DataFrame({
    "city": ["Dhaka","Chittagong","Khulna","Rajshahi"],
    "temperature": [37,45,37,45],
})
df1
```

```python
df2= pd.DataFrame({
    "city": ["Dhaka","Chittagong","Sylhet"],
    "humidity": [80, 60, 78]
})
df2
```

```python
df3 = pd.merge(df1,df2, on ='city' , how = 'inner') #common city will print
df3
```

```python
df3 = pd.merge(df1,df2, on ='city' ) #by default database is inner join
df3
```

```python
df3 = pd.merge(df1,df2, on ='city' , how = 'outer') #Both common and uncommon will print
df3
```

```python
df3 = pd.merge(df1,df2, on ='city' , how = 'left') #left/1st dataframe will print
df3
```

```python
df3 = pd.merge(df1,df2, on ='city' , how = 'right') #right/2nd dataframe will print
df3
```

```python
#indicator flag
df3 = pd.merge(df1,df2, on ='city' , how = 'outer',indicator = True) #to know the origin of value. by default it is False
df3
```

```python
#suffixes
df1 = pd.DataFrame({
    "city": ["Dhaka","Chittagong","Khulna","Rajshahi"],
    "temperature": [37,45,37,45],
    "humidity": [80, 60, 78,80]
})
df1
```
